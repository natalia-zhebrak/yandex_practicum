## Проверка гипотез ##

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые ###

**Нулевая гипотеза:** средние пользовательские рейтинги платформ Xbox One и PC не отличаются (разница статистически не значима)


**Альтернативная гипотеза:** средние пользовательские рейтинги платформ Xbox One и PC отличаются (разница статистически значима)

```python
sample_1 = data_actual.loc[(data_actual['platform'] == 'XOne'), 'user_score']
sample_2 = data_actual.loc[(data_actual['platform'] == 'PC'), 'user_score']

alpha = .05

results = st.ttest_ind(
    sample_1, 
    sample_2)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

```python
sample_1.isna().mean()
```

```python
sample_2.isna().mean()
```

В выборках есть пропуски, поэтому p-value не удается рассчитать. Заменим пропуски на -1 и для проверки гипотез отфильтруем только положительные значения

```python
data_actual['user_score'] = data_actual['user_score'].fillna(-1)
```

```python
sample_1 = data_actual.query('user_score >=0').loc[(data_actual['platform'] == 'XOne'), 'user_score']
sample_2 = data_actual.query('user_score >=0').loc[(data_actual['platform'] == 'PC'), 'user_score']

alpha = .05

results = st.ttest_ind(sample_1, sample_2, equal_var = False)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```
